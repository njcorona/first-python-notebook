At this point I start analyzing only the data that concerns Proposition 64, the legalization of marijuana.

```python
# All of the elements in props such that the prop_name is the provided string.
prop64 = props[props.prop_name == 'PROPOSITION 064- MARIJUANA LEGALIZATION. INITIATIVE STATUTE.']  
```

```python
prop64.head()
```

```python
prop64.info()
```

```python
# Here note that both contribs and props have common data:  calaccess_prop_id
contribs.info()
```

```python
# the merge method will return only those rows where a common identifier found in both tables
merged64 = pd.merge(prop64, contribs, on="calaccess_committee_id")
```

```python
merged64.head()
```

```python
merged64.info()
```

```python
# the total amount spent on this proposition
merged64.amount.sum()
```

Here I decide to answer the question: what proportion of contributors are from out of state?

```python
contributorlist64 = merged64.groupby(["contributor_firstname", "contributor_lastname", "committee_position", "contributor_state"]).amount.sum().reset_index().sort_values("amount", ascending=False)
```

```python
outofstatecontributors64 = contributorlist64[contributorlist64.contributor_state != 'CA'] # contributorlist64outofstate = 
```

```python
# Percentage of contributors who are from out of state
(float(len(outofstatecontributors64.index))) / (len(contributorlist64.index))
```

Here I ask:  who are the employers of the top opposers and supporters of Proposition 64?

```python
contributorlistwemployers64 = merged64.groupby(["contributor_firstname", "contributor_lastname", "committee_position", "contributor_employer"]).amount.sum().reset_index().sort_values("amount", ascending=False)
```

```python
# Top supporters' employers
contributorlistwemployers64[contributorlistwemployers64.committee_position == "SUPPORT"].head()
```

```python
# Top opposers' employers
contributorlistwemployers64[contributorlistwemployers64.committee_position != "SUPPORT"].head()
```

```python
# Opposer employers
contributorlistwemployers64[contributorlistwemployers64.committee_position != "SUPPORT"].contributor_employer.value_counts().reset_index()
```

```python
# Supporter employers
contributorlistwemployers64[contributorlistwemployers64.committee_position == "SUPPORT"].contributor_employer.value_counts().reset_index()
```
